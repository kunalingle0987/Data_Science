```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_csv('sales_data.csv')
df['Date']=pd.to_datetime(df['Date'])
```

```python
# print(df.head())
# print(df.info())
# print(df.describe())
```

```python
# Sales Over Time

sales_over_time=df.groupby(df["Date"].dt.to_period("M")).Total_Sales.sum()

sales_over_time.plot(kind='bar',figsize=(10,6))
plt.title("Monthly_Sales")
plt.xlabel('Months')
plt.ylabel('total_Sales')
plt.show()
```

```python
#top_Products
```

```python
top_product=df.groupby('Product').Total_Sales.sum().sort_values(ascending=False).head(5)
```

```python
top_product.plot(kind='bar', color='skyblue')
plt.title("Top 5 Product by Sales")
plt.ylabel("Total_Sales")
plt.show()
```

```python
#Sales By Region and Product
pivot= df.pivot_table(index="Region", columns="Product", values="Total_Sales", aggfunc="sum", fill_value=0)
sns.heatmap(pivot,annot=True,fmt=".1f",cmap='Blues')
plt.title('Sales_Heatmap by region & Product')
plt.show()
```
